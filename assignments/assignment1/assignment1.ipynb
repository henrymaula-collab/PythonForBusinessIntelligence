## Task 1 

Filter invalid transactions, compute totals, group by region, and find the top region.

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

# a) Remove transactions with zero price or negative quantity
valid_transactions = [
    t for t in transactions
    if t["price"] > 0 and t["quantity"] > 0
]

# b) Add a "total" field to each valid transaction
for t in valid_transactions:
    t["total"] = t["price"] * t["quantity"]

# c) Group by region and sum revenue
revenue_by_region = {}
for t in valid_transactions:
    region = t["region"]
    revenue_by_region[region] = revenue_by_region.get(region, 0) + t["total"]

print("Revenue by region:", revenue_by_region)

# d) Region with the highest total revenue
top_region = max(revenue_by_region, key=revenue_by_region.get)
print(f"Highest revenue region: {top_region} with €{revenue_by_region[top_region]}")
```

## Task 2 

Write `summarise_sales` to compute key statistics from a list of daily sales figures.

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]


def summarise_sales(daily_sales):
    """Return a dict with total, average, best/worst day, and above-average count."""
    total = sum(daily_sales)
    average = round(total / len(daily_sales), 2)
    best_day = daily_sales.index(max(daily_sales)) + 1
    worst_day = daily_sales.index(min(daily_sales)) + 1
    above_average_count = sum(1 for s in daily_sales if s > average)

    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count,
    }


summary = summarise_sales(sales)
print(summary)
```

## Task 3 

Use loops and conditionals to analyse student records.

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

# a) Students with a passing grade (>= 50)
passing = []
for student in students:
    if student["grade"] >= 50:
        passing.append(student["name"])

print("Passing students:", passing)

# b) Status message per student
print()
for student in students:
    name = student["name"]
    grade = student["grade"]
    if grade >= 80:
        print(f"{name}: Excellent")
    elif grade >= 50:
        print(f"{name}: Pass")
    else:
        print(f"{name}: Fail")

# c) Count students with attendance below 80 %
low_attendance_count = 0
for student in students:
    if student["attendance"] < 0.80:
        low_attendance_count += 1

print(f"\nStudents with attendance below 80%: {low_attendance_count}")

# d) Highest grade using a manual loop (no max())
top_student = students[0]
for student in students[1:]:
    if student["grade"] > top_student["grade"]:
        top_student = student

print(f"Top student (loop): {top_student['name']} with grade {top_student['grade']}")

# e) Highest grade using max()
top_student_max = max(students, key=lambda s: s["grade"])
print(f"Top student (max):  {top_student_max['name']} with grade {top_student_max['grade']}")
```

## Task 4 

Design a `Portfolio` class with methods for managing investment holdings.

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

    # a) Add or update an asset
    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings:
            self.holdings[ticker]["shares"] += shares
            self.holdings[ticker]["price_per_share"] = price_per_share
        else:
            self.holdings[ticker] = {
                "shares": shares,
                "price_per_share": price_per_share,
            }

    # b) Total market value of the portfolio
    def total_value(self):
        return sum(
            h["shares"] * h["price_per_share"]
            for h in self.holdings.values()
        )

    # c) Ticker with the highest total value
    def most_valuable(self):
        return max(
            self.holdings,
            key=lambda t: self.holdings[t]["shares"] * self.holdings[t]["price_per_share"],
        )

    # d) Human-readable summary
    def __str__(self):
        return (
            f"Portfolio of {self.owner}: "
            f"{len(self.holdings)} asset(s), "
            f"total value €{self.total_value():,.2f}"
        )


# --- Demonstration ---
portfolio = Portfolio("Henry")
portfolio.add_asset("AAPL", 10, 185.50)
portfolio.add_asset("MSFT", 5, 420.00)
portfolio.add_asset("NVDA", 8, 950.00)

print(portfolio)
print(f"Total value:    €{portfolio.total_value():,.2f}")
print(f"Most valuable:  {portfolio.most_valuable()}")
```

## Task 5 

Parse order dates with the `datetime` module and compute the date span.

```python
from datetime import datetime

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"),  ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

order_dates = [
    datetime.strptime(date_str, "%Y-%m-%d").date()
    for date_str, _ in orders
]

earliest = min(order_dates)
latest = max(order_dates)
span = (latest - earliest).days

print(f"Earliest order: {earliest}")
print(f"Latest order:   {latest}")
print(f"Days between:   {span}")
```

## Task 6 

Use NumPy to analyse quarterly revenue data.

```python
import numpy as np

revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310],
])

# a) Total annual revenue per year (row sums)
annual_revenue = revenue.sum(axis=1)
print("Annual revenue per year:", annual_revenue)

# b) Mean quarterly revenue across all years (column means)
quarterly_means = revenue.mean(axis=0)
print("Mean quarterly revenue:", quarterly_means)

# c) Apply growth adjustment to the 2024 row via broadcasting
growth_factors = np.array([1.05, 1.03, 1.08, 1.02])
adjusted_2024 = revenue[2] * growth_factors
print("Adjusted 2024 revenue:", adjusted_2024)

# d) All quarterly values exceeding 250k
above_250 = revenue[revenue > 250]
print("Quarters above 250k:", above_250)
```

## Task 7 

Manipulate an employee DataFrame with filtering, sorting, and column operations.

```python
import pandas as pd

data = {
    "employee": ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department": ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary": [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9],
}

df = pd.DataFrame(data)

# a) Salary efficiency: salary per year of experience
df["salary_per_year_exp"] = (df["salary"] / df["years_exp"]).round(0).astype(int)
print("With salary_per_year_exp column:")
print(df, "\n")

# b) Top 3 employees by performance score
top3 = df.sort_values("performance_score", ascending=False).head(3)
print("Top 3 by performance:")
print(top3, "\n")

# c) Finance employees with performance score > 4.0
finance_high = df.loc[
    (df["department"] == "Finance") & (df["performance_score"] > 4.0)
]
print("Finance & score > 4.0:")
print(finance_high, "\n")

# d) Rename years_exp -> experience_years
df = df.rename(columns={"years_exp": "experience_years"})
print("After rename:")
print(df.columns.tolist(), "\n")

# e) Salary and performance score for the first four employees
subset = df.iloc[:4][["salary", "performance_score"]]
print("First 4 employees (salary & performance):")
print(subset)
```

## Task 8 

Generate website traffic data and compute derived metrics.

```python
import numpy as np
import pandas as pd

# Generate the data (provided seed ensures reproducibility)
rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

# a) Build the DataFrame with date as the index
traffic = pd.DataFrame({
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value,
}, index=days)
traffic.index.name = "date"

print(traffic.head(), "\n")

# b) Derived columns
traffic["conversions"] = (traffic["visitors"] * traffic["conversion_rate"]).round(0).astype(int)
traffic["daily_revenue"] = traffic["conversions"] * traffic["avg_order_value"]

print(traffic.head())
```
