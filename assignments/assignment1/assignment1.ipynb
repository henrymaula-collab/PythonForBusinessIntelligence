## Task 2 — Creating a Function

Write `summarise_sales` to compute key statistics from a list of daily sales figures.

Task 2: Creating a function (1p)

You are given the following list of daily sales figures (in euros) for a two-week period:

sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
Write a function called summarise_sales that takes a list of daily sales values and returns a dictionary containing:

"total": the sum of all daily sales

"average": the mean daily sales value, rounded to two decimals

"best_day": the index (starting from 1) of the day with the highest sales

"worst_day": the index (starting from 1) of the day with the lowest sales

"above_average_count": the number of days where sales exceeded the average

Call the function with the list above and print the returned dictionary.

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]


def summarise_sales(daily_sales):
    #Returns a dict with total, average, best/worst day, and above-average count.
    total = sum(daily_sales)
    average = total / len(daily_sales) #earlier roundede round(...,2)
    best_day = daily_sales.index(max(daily_sales)) + 1
    worst_day = daily_sales.index(min(daily_sales)) + 1
    above_average_count = sum(1 for s in daily_sales if s > average)

    return dict(
        total = total,
        average = average,
        best_day = best_day,
        worst_day = worst_day,
        above_average_count = above_average_count,
    )


summary = summarise_sales(sales)
print(summary)
```

## Task 4 — Class and Methods

Design a `Portfolio` class with methods for managing investment holdings.

Task 4: Class and methods (1p)

Design a class called Portfolio that represents an investment portfolio. The constructor is provided for you:

class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}
Each key in holdings is a ticker symbol (string) and each value is a dictionary with keys "shares" and "price_per_share". Add the following methods to the class:

a) A method add_asset(ticker, shares, price_per_share) that adds a new asset or updates the number of shares if the ticker already exists (in which case the price should also be updated).

b) A method total_value() that returns the total market value of the portfolio.

c) A method most_valuable() that returns the ticker of the asset with the highest total value (shares × price).

d) A __str__ method that returns a readable summary of the portfolio, including the owner's name, the number of assets, and the total value.

Create an instance, add at least three assets, and demonstrate all methods with printed output.

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

    # a) Add or update an asset
    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings:
            self.holdings[ticker]["shares"] += shares
            self.holdings[ticker]["price_per_share"] = price_per_share
        else:
            self.holdings[ticker] = {
                "shares": shares,
                "price_per_share": price_per_share,
            }

    # b) Total market value of the portfolio
    def total_value(self):
        return sum(
            h["shares"] * h["price_per_share"]
            for h in self.holdings.values()
        )

    # c) Ticker with the highest total value
    def most_valuable(self):
        return max(
            self.holdings,
            key=lambda t: self.holdings[t]["shares"] * self.holdings[t]["price_per_share"],
        )

    # d) Readable summary
    def __str__(self):
        return (
            f"Portfolio of {self.owner}: "
            f"{len(self.holdings)} asset(s), "
            f"total value €{self.total_value():,.2f}"
        )


# --- Demonstration ---
portfolio = Portfolio("Warren Buffet")
portfolio.add_asset("AAPL", 10, 185.50)
portfolio.add_asset("MSFT", 5, 420.00)
portfolio.add_asset("NVDA", 8, 950.00)

print(portfolio)
print(f"Total value:    €{portfolio.total_value():,.2f}")
print(f"Most valuable:  {portfolio.most_valuable()}")
```

## Task 5 — Importing a Module

Parse order dates with the `datetime` module and compute the date span.

Task 5: Importing a module (1p)

You are given the following list of order dates (as strings) and product names:

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]
Using the datetime module, parse each date string into a datetime.date object. Compute and print the number of days between the earliest and latest order.

```python
from datetime import datetime

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"),  ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

order_dates = [
    datetime.strptime(date_str, "%Y-%m-%d").date()
    for date_str, _ in orders
]

earliest = min(order_dates)
latest = max(order_dates)
span = (latest - earliest).days

print(f"Earliest order: {earliest}")
print(f"Latest order:   {latest}")
print(f"Days between:   {span}")
```

## Task 6 — Array Operations and Broadcasting

Use NumPy to analyse quarterly revenue data.

Task 6: Array Operations and Broadcasting (1p)

A company tracks quarterly revenue (in thousands of euros) across four quarters over three years:

# Rows: years (2022, 2023, 2024), Columns: quarters (Q1, Q2, Q3, Q4)
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]])
a) Compute and print the total annual revenue for each year (i.e., row sums). Use the axis parameter.

b) Compute the mean quarterly revenue across all years (i.e., column means).

c) Using broadcasting, apply a growth adjustment to the 2024 row by multiplying it element-wise with the array [1.05, 1.03, 1.08, 1.02]. Print the adjusted values.

d) Use boolean indexing to find and print all quarterly revenue values that exceed 250 (thousands of euros).

```python
import numpy as np

revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310],
])

# a) Total annual revenue per year (row sums)
annual_revenue = revenue.sum(axis=1)
print("Annual revenue per year:", annual_revenue)

# b) Mean quarterly revenue across all years (column means)
quarterly_means = revenue.mean(axis=0)
print("Mean quarterly revenue:", quarterly_means)

# c) Apply growth adjustment to the 2024 row via broadcasting
growth_factors = np.array([1.05, 1.03, 1.08, 1.02])
adjusted_2024 = revenue[2] * growth_factors
print("Adjusted 2024 revenue:", adjusted_2024)

# d) All quarterly values exceeding 250k
above_250 = revenue[revenue > 250]
print("Quarters above 250k:", above_250)
```

    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9]

a) Add a new column salary_per_year_exp computed as salary / years_exp. Round to the nearest integer.

b) Sort the DataFrame by performance_score in descending order and print the top 3 employees.

c) Using .loc, select all employees in the Finance department with a performance score above 4.0. Print the result.

d) Rename the column years_exp to experience_years using the .rename() method.

e) Using .iloc, select and print the salary and performance score columns for the first four employees.

```python
import pandas as pd

data = {
    "employee": ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department": ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary": [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9],
}

df = pd.DataFrame(data)

# a) Salary efficiency -> salary per year of experience
df["salary_per_year_exp"] = (df["salary"] / df["years_exp"]).round(0).astype(int)
print("With salary_per_year_exp column:")
print(df, "\n")

# b) Top 3 employees by performance score
top3 = df.sort_values("performance_score", ascending=False).head(3)
print("Top 3 by performance:")
print(top3, "\n")

# c) Finance employees with performance score > 4.0
finance_high = df.loc[
    (df["department"] == "Finance") & (df["performance_score"] > 4.0)
]
print("Finance & score > 4.0:")
print(finance_high, "\n")

# d) Rename years_exp -> experience_years
df = df.rename(columns={"years_exp": "experience_years"})
print("After rename:")
print(df.columns.tolist(), "\n")

# e) Salary and performance score for the first four employees
subset = df.iloc[:4][["salary", "performance_score"]]
print("First 4 employees (salary & performance):")
print(subset)
```

Using this generated data, perform the following:

a) Create a DataFrame with columns: date, visitors, conversion_rate, and avg_order_value. Set date as the index.

b) Compute two new columns: conversions (= visitors × conversion_rate, rounded to the nearest integer) and daily_revenue (= conversions × avg_order_value).

```python
import numpy as np
import pandas as pd

# Generate the data (provided seed ensures reproducibility)
rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

# a) Build the DataFrame with date as the index
traffic = pd.DataFrame({
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value,
}, index=days)
traffic.index.name = "date"

print(traffic.head(), "\n")

# b) Derived columns
traffic["conversions"] = (traffic["visitors"] * traffic["conversion_rate"]).round(0).astype(int)
traffic["daily_revenue"] = traffic["conversions"] * traffic["avg_order_value"]

print(traffic.head())
```
